# Análise explorátória dos dados

```python
#Importando a biblioteca e renomeando era para 'pd'
import pandas as pd
```

```python
#Trazendo a nossa base de dados do github e armazenando ela em uma variável chamada 'url'.
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

```python
dados.head(10)
```

```python
#Média de todos os valores da coluna "Valor"
dados['Valor'].mean()
```

```python
#Calculando a média unicamente nas colunas numéricas
dados.groupby('Tipo').mean(numeric_only = True)
```

```python
#Especificando a coluna numérica onde queremos aplicar a média
dados.groupby('Tipo')['Valor'].mean()
```

```python
#Ordenação das informações
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14,10), color = 'black');
```

#Removendo os imóveis comerciais

```python
#Pegando os valores únicos da coluna 'Tipo'
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém','Casa Comercial','Terreno Padrão',
                      'Loja Shopping/ Ct Comercial', 'Box/Garagem','Loft',
                      ]
```

```python
#Query nos permite selecionar linhas de um dataframe com base em uma condição específica
dados.query('@imoveis_comerciais in Tipo')
#Neste caso, estamos selecionando os valores da variável "imoveis_comerciais" que também estão presentes na coluna "Tipo"
```

```python
#Query nos permite selecionar linhas de um dataframe com base em uma condição específica
df = dados.query('@imoveis_comerciais not in Tipo')
df.head()
```

```python
df.Tipo.unique()
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14,10), color = 'black');
```

```python
#Selecionando imóveis com uma ou mais suítes
dados.query('Suites >= 1')
```

```python
#Contar quantas vezes os valores da coluna 'Tipo' se repetem
df.Tipo.value_counts()
```

```python
#Contar quantas vezes os valores da coluna 'Tipo' se repetem no formato de porcentual
df.Tipo.value_counts(normalize = True)
```

```python
#Converte uma series em um dataframe
df.Tipo.value_counts(normalize = True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo = df.Tipo.value_counts(normalize = True).to_frame().sort_values('Tipo')
df_percentual_tipo.plot(kind='bar', figsize=(14,10), color='black',
                        xlabel = ' Tipos', ylabel = 'Percentual')
```

```python
#Realizando uma query que lista todos os apartamentos da coluna 'Tipo'
df.query('Tipo == "Apartamento"')
```

```python
#Alterando o conteúdo da variável 'df'
df = df.query('Tipo == "Apartamento"')
df.head()
```
