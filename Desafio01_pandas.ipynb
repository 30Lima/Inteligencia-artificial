```python
#importando a lib para realizar a análise de daos
import pandas as pd
```

```python
#Guardando o dataset em uma variável
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url)
```

```python
#Calculando a média aritmética das notas e idade
media_aritmetica_notas = dados['Notas'].mean()
print("A média aritmética de notas é: ",media_aritmetica_notas)

media_aritmetica_idade = dados['Idade'].mean()
print("A média aritmética das idades é:", media_aritmetica_idade)
```

```python
#Calculando o desvio padrão das notas e idade
desvio_padrao_idades = dados['Idade'].std()
print("O desvio padrão das idades é: ", desvio_padrao_idades)

desvio_padrao_notas = dados['Notas'].std()
print("O desvio padrão de notas é: ", desvio_padrao_notas)
```

```python
#Calculando o valores mínimo e máximo de notas e idade para ter noção da amplitude do dataframe
valor_minimo = dados['Notas'].min()
print("O valor mínimo do dataframe é: ", valor_minimo)

valor_maximo = dados['Notas'].max()
print("O valor máximo do dataframe é: ", valor_maximo)
```

```python
#Calculando a distribuição de valores das notas e idade
distribuicao_valores_notas = dados['Notas'].quantile()
print("A distribuição de notas é: ", distribuicao_valores_notas)

distribuicao_valores_idade = dados['Idade'].quantile()
print("A distribuição de idades é: ", distribuicao_valores_idade)
```
