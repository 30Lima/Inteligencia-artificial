```python
#Importando a biblioteca pandas e renomeando ela para 'pd'.
import pandas as pd
```

```python
#Trazendo a nossa base de dados do github e armazenando ela em uma variável chamada 'url'.
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
#Lendo a nossa base de dados e mostrando na tela.
pd.read_csv(url)
```

```python
#Mostrando a base de dados novamente e aplicando uma melhor formatação para análise
dados = pd.read_csv(url, sep=';')
#Verificando se a variável foi salva corretamente
```

```python
#Exibindo as 10 primeiras linhas da nossa base de dados (a contagem sempre se inicia a partir do zero)
dados.head(10)
```

```python
#Exibindo as 10 últimas linhas da nossa base de
dados.tail(10)
```

Para uma melhor análise, é recomendado que utilize o "head" e "tail".
Eles ajudam a realizar uma análise exploratória sólida e menor confusa.
